## Part 2: HEART Score

The HEART score is a predictor for major cardiac events.  It requires 5 high-level inputs:
* History (Slightly / Moderately / Highly suspicious)
* EKG (Normal / Non-specific repolarization disturbance / Significant ST deviation)
* Age (Number)
* Risk Factors (Number of risk factors)
* Initial Troponin (Number of times the normal limit)

Fill out the function below with logic to calculate the numeric risk score for teh given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation. Also make sure your code satisfies the test cases provided in the assert statements.

````python
def heart(history, ekg, age, risks, troponin):
    """(str, str, int,str, float) History catogorizes the presentation as Slightly / Moderately / Highly suspicious,. 
    EKG categorizes the ekg findings into normal, Non-specific repolarization disturbance / Significant ST deviation.
    age is age at presentation
    Risk factors indicate the number of risk factors a patient has. 
    Initial troponin defines as number of times greater then the normal limit.
    This function gives a score which predicts major cardiac events
    >>> heart('Highly suspicious', 'Normal', 66, 3, .5)
    6
    """
    if history== 'Moderately suspicious':
        hx= 1
    elif history =='Highly suspicious':
        hx=2
    elif history=='Slightly suspicious':
        hx=0
    if ekg=='Non-specific repolarization':
        ekgS=1
    elif ekg=='Significant ST deviation':
        ekgS=2
    else:
        ekgS=0
    if age>=65:
        age2= 2
    elif age<45:
        age2=0
    else:
        age2=1
    if risks==0:
        rf=0
    elif risks>2:
        rf=2
    else:
        rf=1
    if troponin>3:
        trop=2
    elif troponin<=1:
        trop=0
    else:
        trop=1
    heart=hx+ekgS+age2+rf+trop
    #print(hx, ekgS, age2, rf, trop, heart)
    return heart
    

    
    
    
        
    
    ### YOUR SOLUTION HERE
````

## Part 3: Framingham Risk Score for Hard Coronary Heart Disease

The Framingham Risk Score for Hard Coronary Heart Disease is intended for non diabetic patients age 30-79 only.  So, if the patient's age is < 30 or > 79, your function should return `-1` rather than a specific risk score.

The Framingham Risk Score takes 7 inputs:
* Age (Number)
* Sex (Male / Female)
* Smoker (True / False)
* Total cholesterol (Number)
* HDL cholesterol (Number)
* Systolic BP (Number)
* Blood pressure being treated with medicines (No / Yes)

You'll not that rather than being a basic parametric equation, this is a regression function defined by coefficients.  It also requires you take the natural logarithm (`ln`) of many of the parameters.  To help you out, here's an example of how to interpret the formulat provided in the website's **Evidence** tab.

Take special note of the footnotes in the logic:
* *Yes=1, No=0 (for Treated for blood pressure and Smoker)
* ** Men: if age >70, use ln(70) x Smoker. Women: if age >78, use ln(78) x Smoker.


---

These segments of the equation and the coefficient table...

> $ L_{Men} = \beta \times \ln(Age) + \beta \times \ln(cholesterol) ... $
>
> | Variable &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | Men | Women
> | :---------------------------- | :--: | :--: |
> | $ \ln(Age) $ | 52.00961 | 31.764001
> | $ \ln(Cholesterol) $ | 20.014077 | 22.465206

Could be written in Python as follows:

````python
import math
def framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated):
    """(int, str, bool, float, float, float, bool)-> float
    age is age at presentation, if age is greater than 79 or less than 30 a -1 will be returned. 
    sex is either male or femmale, smoker is defined as current smoker, cholesterol is total cholesterol, hdl is hdl cholesterol
    systolic is systolic blood pressure and bp_treated indicated if a patient is taking medication for hypertension. 
     >>> framingham (54,'Male',False,156,52,117,True)
     0.0437
    """
    ### YOUR SOLUTION HERE
    if age > 79:
        p=-1
    elif age <30:
        p=-1
    else:
        SubT = 0
        if sex == "Male":
            A = 52.00961 * math.log(age)
            C = 20.014077 * math.log(cholesterol)
            H = -0.905964 * math.log(hdl)
            S = 1.305784 * math.log(systolic)
            AC = -4.605038* math.log(age) *math.log(cholesterol)
            A2= -2.93323* math.log(age)* math.log(age)
            if bp_treated==True:
                tx = 0.241549
            else:
                tx=0
                #S=1.305784 * math.log(systolic)
            if smoker==False:
                sm=0
                SA=0
            else:
                sm=12.096316
                if age >70:
                    SA=-2.84367* math.log(70)
                else:
                    SA=-2.84367*math.log(age)
            SubT= A+ C+ H+ S+ sm+ AC+ tx+ SA+ A2-172.300168
            T=1-(.9402** math.exp(SubT))
            p=round(T,4)
            #print (A, C, H, S, Tx, sm, AC, SA, A2, T, p)
            #print (p)
        else:
            A = 31.764001 * math.log(age)
            C = 22.465206 * math.log(cholesterol)
            H = -1.187731* math.log(hdl)
            S = 2.552905 *math.log(systolic)
            AC = -5.060998* math.log(age) *math.log(cholesterol)
            if bp_treated==True:
                tx = 0.420251
            else:
                #S= 2.552905 * math.log(systolic)
                tx=0
                #tx= 2.552905 *math.log(systolic)
            if smoker==False:
                sm=0
                SA=0
            else:
                sm=13.07543
                if age >78:
                    SA= -2.996945* math.log(78)
                elif age<78:
                    SA= -2.996945* math.log(age)
            SubT= A+ C+ H+ tx+S+ sm+ AC+ SA- 146.5933061
            T = 1 - (0.98767**math.exp(SubT))
            p =round(T,4)
           # print (p)
    return p
````
